```python
import matplotlib
import matplotlib.pyplot as plt 
import numpy as np 

barWidth = 0.25

nyc = [185111028, 245809210, 278151116, 168402018, 179130258]
chi = [22509331.30, 31296262.87, 94623598.75, 56950918.16, 32190159.51]
la = [37260955, 77619184, 42919819, 26335683, 32116659]

r1 = np.arange(len(nyc))
r2 = [x + barWidth for x in r1]
r3 = [x + barWidth for x in r2]

fig, ax = plt.subplots()
ax.ticklabel_format(style='plain')

ax.set_ylabel('Settlement Amount', fontweight='bold')
ax.set_title('Settlement Amount by Year and Cities')

plt.bar(r1, nyc, color='#A271E8', width=barWidth, edgecolor='white', label='nyc')
plt.bar(r2, chi, color='#33EBFF', width=barWidth, edgecolor='white', label='chicago')
plt.bar(r3, la, color='#33FF89', width=barWidth, edgecolor='white', label='la')

plt.xlabel('year', fontweight='bold')
plt.xticks([r + barWidth for r in range(len(nyc))], ['2015', '2016', '2017', '2018', '2019'])

axes = plt.gca()
axes.set_ylim([0, 300000000])

plt.legend()
plt.show()
```
