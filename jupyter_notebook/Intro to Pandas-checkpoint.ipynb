# Pandas 2 Challenge: 

Which geography has the highest median income? 


```python
df.sort_values('median_hh_income', ascending=False).head()
```

```python
df[df.median_hh_income == df.median_hh_income.max()]
```

What large geography (1MM+) has the smallest average family size? 


```python
df[df.population > 1000000].sort_values('avg_family_size').head()
```

Which geographies in NY have the highest and lowest median age?

```python
ny_df = df[df.geography.str.contains('NY')]
```

```python
ny_df.sort_values('median_age', ascending=False).head() # Highest
```

```python
ny_df.sort_values('median_age').head() # Lowest
```
