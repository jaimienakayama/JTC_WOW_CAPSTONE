```python
import pandas as pd
```

```python
# Read the csv file
df = pd.read_csv('census.csv', index_col='FIPS')
```

```python
df.columns = ['geography', 'population', 'median_age', 
              'total_households', 'median_hh_income',
             'avg_hh_size', 'total_families', 'avg_family_size',
             'pct_owner_occupied_housing', 'pct_renter_occupied_housing',
             'pop_16_and_over', 'unemployment_rate']
```

# Run linear regression

```python
from statsmodels import regression
import scipy as sp
import numpy as np
```

```python
x = df [['median_age', 'total_families']]
y = df.total_households
```

```python
# add a constant
x['intercept'] = 1
x.head()
```

```python
# creat model (OLS = ordinary least squares)
model = regression.linear_model.OLS(y,x)

# fit the model
results = model.fit()

#look at model results
results.summary()
```
