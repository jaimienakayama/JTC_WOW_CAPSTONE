```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
def float_to_int(x):
    split_string = x.split(".", 1)
    substring = split_string[0]
    if len(substring) > 0: 
        return int(substring) 
    else: 
        return ''

all_cities_full = pd.read_csv('jaimie.csv', converters={'calendar_year': float_to_int})
all_cities_full
```

```python
boston_filter = all_cities_full[all_cities_full.city == "Boston"]
```

```python
boston_filter
```

```python
boston_columns = boston_filter.loc[:, ['city', 'calendar_year', 'amount_awarded']]
```

```python
boston_columns
```

```python
boston_graph_mean = boston_filter.groupby(['calendar_year'])[['amount_awarded']].mean()
```

```python
boston_graph_mean.round()
```

```python
boston_graph_sum = boston_filter.groupby(['calendar_year'])[['amount_awarded']].sum()
```

```python
boston_graph_sum
```

```python
plot = boston_graph_mean.plot.bar()
plot.yaxis.set_major_formatter(mpl.ticker.StrMethodFormatter('{x:,.0f}'))
plot.set_title('Boston City Average Payout For Police Misconduct By Year')
plot.set_xlabel('Year')
plot.set_ylabel('Amount Awarded')
```

```python
boston_graph = boston_graph_mean.plot()
boston_graph.yaxis.set_major_formatter(mpl.ticker.StrMethodFormatter('{x:,.0f}'))
boston_graph.set_title('Boston City Average Payout For Police Misconduct By Year')
boston_graph.set_xlabel('Year')
boston_graph.set_ylabel('Amount Awarded')

```

```python
plot = boston_graph_mean.plot.barh()
plt.gcf().axes[0].xaxis.get_major_formatter().set_scientific(False)
plot.xaxis.set_major_formatter(mpl.ticker.StrMethodFormatter('{x:,.0f}'))
plot.set_title('Boston City Average Payout For Police Misconduct By Year')
plot.set_xlabel('Amount Awarded')
plot.set_ylabel('Year')
```
