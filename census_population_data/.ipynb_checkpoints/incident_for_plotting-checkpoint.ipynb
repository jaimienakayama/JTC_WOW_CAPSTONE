```python
import pandas as pd

```

```python
df = pd.read_csv('unfinished.csv') # inorder to create a new csv that contains the total amount of cases per state, i used an old that contained raw combinded data 
```

```python
df
```

```python
 yearly = df.groupby(['city','calendar_year'])['calendar_year'].count()  #grouping to get yearly incident amount per city
```

```python
yearly #returns a series
```

```python
yearly = pd.Series(yearly)
yearly =yearly.to_frame()     #converting series of incidents yearly into DataFrame
```

```python
yearly
```

```python
yearly.columns 
```

```python
yearly.index #Note the index is a multi-index with year and city 
```

```python
yearly.rename(columns ={'calendar_year':'yearly_incidents'}, inplace=True) #renaming column 
```

```python
yearly =yearly.reset_index(0).reset_index(drop=False) #reseting index to seperate multi-indexed year as column
```

```python
yearly
```

```python
yearly['total'] = yearly[['city','yearly_incidents']].groupby('city').transform('sum') 
 
    #creating new "total" column for the toal amount of incidents per city
```

```python
yearly
```

```python
yearly.to_csv('incidents_data.csv')
```
