## Importing Pandas library

```python
import pandas as pd
```

## Reading excel format data and assigning it to a variable

```python
df = pd.DataFrame(pd.read_excel('SUB-IP-EST2019-ANNRNK.xlsx.xlsx')) # reading excel format data
```

### Displaying data

```python
df
```

## Setting columns and dropping NaN values

```python
df.columns=['id','Geographic_Area','Census','Estimates Base','2010',
                   '2011','2012','2013','2014','2015','2016','2017','2018','2019'] #setting columns
df.dropna(inplace=True) #dropping empty 
```

## Setting index, sorting, and deleting column

```python
df.set_index('id', drop=False, inplace=True) #setting index 
df.sort_index(axis=0,ascending=True, inplace=True) #sorting index 
del df['id']   #del column
```

## Getting city index with a for loop using iterows function

```python
cities = ['New York city, New York', 'Los Angeles city, California', 'Chicago city, Illinois',
          'North Charleston city, South Carolina','Boston city, Massachusetts','Cambridge city, Massachusetts',
          'Charleston city, South Carolina','Fort Lauderdale city, Florida','Richmond city, Virginia',
          'Waterbury city, Connecticut','Springfield city, Massachusetts','St. Louis city, Missouri',
          'Paterson city, New Jersey','Cleveland city, Ohio','Memphis city, Tennessee','San Francisco city, California',
          'Milwaukee city, Wisconsin','Baton Rouge city, Louisiana','Cincinnati city, Ohio',
          'Columbia city, South Carolina','Miami city, Florida','Detroit city, Michigan','New Orleans city, Louisiana',
          'Washington city, District of Columbia','Baltimore city, Maryland','Orlando city, Florida'
          ]
rows = []
for index,data in df.iterrows():              #iterrows function for indexing 
    city_name = data['Geographic_Area']           #getting index
    if city_name in cities:
        rows.append(index)
rows
```

## Using iloc function and assinging cities to a variable

```python
city_pop = df.iloc[[0,1,2,5,15,16,19,20,23,27,30,36,49,52,63,64,70,96,99,125,141,180,199,210,237,252,284,166,320,41]]
        #indexing and selecting data #assigning to a variable
```

```python
city_pop
```

## Using assign function to create a new column with edited 'Geographic_Area' = 'city_name'. Iterating though column and removing 'city' + capitalizing 'City' in New York row

```python
new_city_df =city_pop.assign(
    city_name = 
     lambda dataframe: dataframe['Geographic_Area'].map(lambda Geographic_Area: 'New York City, New York'
     if 'New York' in Geographic_Area else Geographic_Area.replace(' city', '')))
#using assign function() to add a Column that edits all 'Geographic_Area' values

```

```python
new_city_df
```

## Dropping column

```python
new_city_df.drop(['Geographic_Area'],axis=1,inplace=True) #dropping 'Geographic_Area' column
```

## Reordering columns 

```python
new_city_df[['city_name','Census','Estimates Base','2010','2011', '2012',
       '2013','2014','2015','2016','2017','2018','2019']]     
```

## Converting new data frame into a csv

```python
new_city_df.to_csv('population_data.csv')
```
