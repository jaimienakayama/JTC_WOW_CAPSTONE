```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl 
import seaborn as sns

df = pd.read_csv('final.csv')
pd.set_option('display.float_format', lambda x: '%.f' %x)
```

```python
df
```

```python
groupby_city = df.groupby(['city'], as_index= False)['amount_awarded'].sum().round()
```

```python
groupby_city
```

```python
population = pd.read_csv('population_data.csv')
```

```python
population = population.groupby(['city_name'], as_index = False)[['Estimates Base']].sum()
```

```python
population
```

```python
frames = [population, groupby_city]
frames
```

```python
result = pd.concat(frames, axis =1)
result.drop(columns = ['city_name'], inplace=True)
result
```

```python
first_column = result.pop('city')
```

```python
result.insert(0, 'city', first_column)
middle_column = result.pop('amount_awarded')
```

```python
result.insert(1, 'amount_awarded', middle_column)
```

```python
result['per capita'] = result['amount_awarded']/result['Estimates Base']
```

```python
result.drop(29, inplace =True)
```

```python
result
```

```python
result.to_csv('per_cap.csv')
```
