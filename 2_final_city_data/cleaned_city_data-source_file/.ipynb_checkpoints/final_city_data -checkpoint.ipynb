## Importing libraries and csv's

```python
import pandas as pd 
import numpy as np


df = pd.read_csv('compleat_raw_cities.csv')
```

```python
df
```

## Displaying the NaN(Not a Number) values in the data frame

```python
df.isnull()
```

## Isolating all rows with NaN values

```python
bool_series = pd.isnull(df["calendar_year"]) 
```

```python
df[bool_series]
```

##  Deleting the NaN values from data base 

```python
df.dropna(subset=['calendar_year'], inplace=True) # calender_year was choosen because any NaN in that column tipically had NaN in the same row
```
