```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
all_cities_full = pd.read_csv('jaimie.csv')
all_cities_full
```

```python
boston_filter = all_cities_full[all_cities_full.city == "Boston"]
```

```python
boston_filter
```

```python
boston_filter.loc[:, ['city', 'calendar_year', 'amount_awarded']]
```

```python
boston_graph_mean = boston_filter.groupby(['calendar_year'])[['amount_awarded']].mean()
```

```python
boston_graph_mean.round()
```

```python
boston_graph_sum = boston_filter.groupby(['calendar_year'])[['amount_awarded']].sum()
```

```python
boston_graph_sum
```

```python
plot = boston_graph_mean.plot.bar()
plt.gcf().axes[0].yaxis.get_major_formatter().set_scientific(False)
plot.set_title('Boston City Average Payout For Police Misconduct By Year')
plot.set_xlabel('Year')
plot.set_ylabel('Amount Awarded')
```

```python
plot = boston_graph_mean.plot()
plt.gcf().axes[0].yaxis.get_major_formatter().set_scientific(False)
plot.set_title('Boston City Average Payout For Police Misconduct By Year')
plot.set_xlabel('Year')
plot.set_ylabel('Amount Awarded')

```

```python
plot = boston_graph_mean.plot.barh()
plt.gcf().axes[0].xaxis.get_major_formatter().set_scientific(False)
plot.set_title('Boston City Average Payout For Police Misconduct By Year')
plot.set_xlabel('Amount Awarded')
plot.set_ylabel('Year')
```
