```python
import pandas as pd  # import pandas library as pd
import numpy as np # import numpy library as np
import matplotlib.pyplot as plt # import matplotlib.pyplot library as plt
import matplotlib as mpl # import matplotlib library as mpl
```

```python
def float_to_int(x): # function to convert float to integer (i.e.: 2014.0 to 2014)
    split_string = x.split(".") # splits input argument at decimal point, returns an array (i.e.: 2014.0 to [2014, 0])
    substring = split_string[0] # takes the value at the index 0 of new array (i.e.: 2014 of [2014, 0])
    if len(substring) > 0: # if the length of the substring is greater then 0 (which means, in the case that there is data in that cell) 
        return int(substring) # convert the substring to an integer, and return that integer
    else: # otherwise (in the case that there is no data in that cell)
        return '' # return an empty string

all_cities_full = pd.read_csv('cleaned.csv', converters={'calendar_year': float_to_int}) # read csv and call the float_to_int function on the column calendar_year
all_cities_full # display csv
```

```python
boston_filter = all_cities_full[all_cities_full.city == "Boston"] # filters through the csv and pulls all the data with the string "Boston" in the city column
boston_filter.loc[:, ['city', 'calendar_year', 'amount_awarded']] # displays only city, calendar_year, and amount_awarded columms
boston_sum = boston_filter.groupby(['calendar_year'])[['amount_awarded']].sum().round() # calculated the sum of the amount awarded for each calendar year and rounded it (used to get flat numbers)
```

```python
plt.rcParams['figure.figsize'] = [7, 6] # controls the size of the graph
boston_line = boston_sum.plot() # plots the data from boston_sum (line graph is the default plot on matplotlib)
boston_line.yaxis.set_major_formatter('${x:,.0f}') # inserts $ value with commas and 0 zeros after the decimal point on the y axis (amount awarded)
boston_line.set_title('Boston: Total Payout For Police Misconduct By Year') # names the title of the graph
boston_line.set_xlabel('Year') # names the x axis
boston_line.set_ylabel('Amount Awarded') # names the y axis
```
